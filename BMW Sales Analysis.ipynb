# BMW Sales Analysis

### Importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python
df=pd.read_csv('BMW sales data (2010-2024).csv')
```

```python
# Group the data by 'Year' and sum the 'Sales_Volume'
sales_by_year = df.groupby('Year')['Sales_Volume'].sum().reset_index()

# Shorten the 'Sales_Volume' numbers to millions and round to 2 decimal places
sales_by_year['Sales_Volume'] = round(sales_by_year['Sales_Volume'] / 1000000, 2)

# Create a line chart with a modified y-axis
plt.figure(figsize=(12, 8))
plt.plot(sales_by_year['Year'], sales_by_year['Sales_Volume'], marker='o', color='#7AB2B2', linestyle='--')

# Set the y-axis to start at 15 million and end at 18 million
plt.ylim(15, 18.5)

# Add data labels to each point
for i, txt in enumerate(sales_by_year['Sales_Volume']):
    plt.annotate(f'{txt:.2f}', (sales_by_year['Year'][i], sales_by_year['Sales_Volume'][i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.title('Total BMW Sales Volume by Year (2010-2024)')
plt.xlabel('Year')
plt.ylabel('Total Sales Volume (in Millions)')
plt.xticks(sales_by_year['Year'], rotation=45)
# plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# Save the plot to a file
plt.savefig('1_sales_volume_by_year.png')
```

```python
# Group the data by 'Year' and 'Region', then sum the 'Sales_Volume'
stacked_sales = df.groupby(['Year', 'Region'])['Sales_Volume'].sum().unstack()

# Shorten the sales numbers to millions and round to 2 decimal places
stacked_sales = round(stacked_sales / 1000000, 2)

# Define the color mapping for each region
# Assumes the regions are in alphabetical order as a default from unstack()
region_colors = {
    'Africa': '#03045e',
    'Asia': '#023e8a',
    'Europe': '#0077b6',
    'Middle East': '#00b4d8',
    'North America': '#90e0ef',
    'South America': '#caf0f8'
}

# Create a list of colors in the correct order for the plot
colors_list = [region_colors.get(region, 'gray') for region in stacked_sales.columns]

# Create a stacked bar chart with the specified colors
plt.figure(figsize=(12, 8))
stacked_sales.plot(kind='bar', stacked=True, figsize=(12, 8), color=colors_list)

# Set the title and labels
plt.title('Total BMW Sales Volume by Year and Region (2010-2024)')
plt.xlabel('Year')
plt.ylabel('Total Sales Volume (in Millions)')
plt.xticks(rotation=45)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
# plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Save the plot to a file
plt.savefig('2_stacked_sales_by_year_and_region.png')
```

```python
# Group the data by 'Year' and 'Fuel_Type', then sum the 'Sales_Volume'
sales_by_year_and_fuel_type = df.groupby(['Year', 'Fuel_Type'])['Sales_Volume'].sum().unstack()

# Shorten the sales numbers to millions and round to 2 decimal places
sales_by_year_and_fuel_type = round(sales_by_year_and_fuel_type / 1000000, 2)

# Define the color mapping as requested by the user
fuel_type_colors = {
    'Petrol': '#00b4d8',
    'Diesel': '#03045e',
    'Hybrid': '#0077b6',
    'Electric': '#023e8a'
}

# Create a list of colors based on the columns of the DataFrame
colors_list = [fuel_type_colors[fuel_type] for fuel_type in sales_by_year_and_fuel_type.columns]

# Create a grouped bar chart with the specified colors
plt.figure(figsize=(12, 8))
sales_by_year_and_fuel_type.plot(kind='bar', figsize=(12, 8), color=colors_list)

# Set the y-axis to start at 3.5 million
plt.ylim(3.5)

# Set the title and labels
plt.title('Total BMW Sales Volume by Year and Fuel Type (2010-2024)')
plt.xlabel('Year')
plt.ylabel('Total Sales Volume (in Millions)')
plt.xticks(rotation=45)
plt.legend(title='Fuel Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Save the plot to a file
plt.savefig('3_sales_by_year_and_fuel_type.png')
```

```python
# Group the data by 'Model' and 'Transmission', then sum the 'Sales_Volume'
sales_by_model_and_transmission = df.groupby(['Model', 'Transmission'])['Sales_Volume'].sum().unstack()

# Shorten the sales numbers to millions and round to 2 decimal places
sales_by_model_and_transmission = round(sales_by_model_and_transmission / 1000000, 2)

# Define a simple color mapping for the transmission types
# Darker blue for Automatic, lighter blue for Manual
transmission_colors = {
    'Automatic': '#003366',
    'Manual': '#336699'
}

# Create a list of colors for the plot based on the column order
colors_list = [transmission_colors[col] for col in sales_by_model_and_transmission.columns]

# Create a grouped bar chart
plt.figure(figsize=(12, 8))
sales_by_model_and_transmission.plot(kind='bar', figsize=(12, 8), color=colors_list)

# Set the y-axis to start at 10 million
plt.ylim(10.5,12.2)

# Set the title and labels
plt.title('Total BMW Sales Volume by Model and Transmission Type')
plt.xlabel('Model')
plt.ylabel('Total Sales Volume (in Millions)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Transmission Type', bbox_to_anchor=(1.05, 1), loc='upper left')
#plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Save the plot to a file
plt.savefig('4_sales_by_model_and_transmission.png')
```
